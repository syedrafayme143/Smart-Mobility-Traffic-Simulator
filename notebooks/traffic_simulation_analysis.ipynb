```python
from sumolib import checkBinary
import os
import traci
import sys

import os
import sys

# Set SUMO_HOME
os.environ['SUMO_HOME'] = "C:/Program Files (x86)/Eclipse/Sumo"  # <- Update if needed

tools = os.path.join(os.environ['SUMO_HOME'], 'tools')
sys.path.append(tools)

def run_simulation_and_get_avg_vehicle_per_edge():
    sumoBinary = checkBinary('sumo-gui')  # or 'sumo' for headless
    traci.start([sumoBinary, "-c", "../simulation/sumo_config.sumocfg"])

    step = 0
    edge_counts = {}

    while traci.simulation.getMinExpectedNumber() > 0:
        traci.simulationStep()

        for edge_id in traci.edge.getIDList():
            vehicle_count = traci.edge.getLastStepVehicleNumber(edge_id)
            if edge_id not in edge_counts:
                edge_counts[edge_id] = []
            edge_counts[edge_id].append(vehicle_count)

        step += 1

    traci.close()

    print("ðŸ“Š Average Vehicles per Edge:")
    for edge_id, counts in edge_counts.items():
        avg = sum(counts) / len(counts)
        print(f"Edge {edge_id}: {avg:.2f} vehicles (avg)")

# Now run the simulation
run_simulation_and_get_avg_vehicle_per_edge()
```
